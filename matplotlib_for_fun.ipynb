## 並列的長條圖

```python
x = arange(1,6)
bar(x - 0.4, [3, 10, 8, 12, 6], width=0.4)
bar(x, [6, 3, 12, 5, 8], width=0.4, color="red")
```

堆起來!!

```python
A = array([3, 10, 8, 12, 6])
B = array([6, 3, 12, 5, 8])
C = array([2, 6, 8, 10, 3])
bar(x, A)
bar(x, B, color="red", bottom = A)
bar(x, C, color="green", bottom = A+B)
```

換個方向

```python
x = arange(1,6)
barh(x - 0.4, [3, 10, 8, 12, 6])
```

比較兩組資料

```python
x = arange(1,6)
A = array([3, 10, 8, 12, 6])
B = array([6, 3, 12, 5, 8])
barh(x - 0.4, A)
barh(x - 0.4, -B, color="red")
```

## 一堆大小不一的圓

## 混合兩個圖

```python
bar(arange(10)-0.5,randint(1,10,10),color='#CCFFCC')
g, = plot(range(10),randint(1,10,10), 'o-', color="#00BFFF", linewidth=5)
g.set_markerfacecolor("white")
g.set_markeredgecolor("#00BFFF")
g.set_markeredgewidth(5)
g.set_markersize(10)
ylim(0,10)
xlim(-0.5,9.5)
```

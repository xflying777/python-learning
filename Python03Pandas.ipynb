## 1. 和亂碼說再見

編碼就是把文字變成數字, 這樣電腦才能「紀錄」下來。

### ASCII 編碼

純英文編碼, 'A' 是 65, 'B' 是 '66' 等。因為英文字母才 26 個, 一個 byte (可記 256) 就足夠。

### Big 5 編碼

用兩個 bytes, 需要包括英文字母, 最多可有 65536 個字, 實際定義只有一萬三千多字。

## 問題

像這種 ASCII 編碼、Big 5 編碼都叫 byte codes, 真正電腦處理尤其存檔一定要轉成某種 byte codes。但以上固定長度的 byte codes, 顯然不足以應付多國語言 (事實上單單中文都有問題) 的需求。

你好 ● こんにちは ● 안녕하세요

## Unicode 來了

Unicode 把「所有人類文字」一一收錄, 不再用 byte 去定義, 而是每個字就給一個號碼。例如:

* 政: \u653F
* 大: \u5927

## UTF-8 編碼

UTF-8 編碼也是一種 byte codes, 每個字的不一定用多少個 bytes。如果是英文的文章, UTF-8 編碼和 ASCII 完全相容, 也就是選用 UTF-8 對英文使用者完全沒有感覺不便之處。

在 Python, 我想把一個字串變成「Unicode 字串」。

```python
name = u"蔡炎龍"
print name
```

```python
print name[0]
```

```python
len(name)
```

比較一下, 就算是可愛的 UTF-8:

```python
name = "蔡炎龍"
print name
```

```python
print name[0]
```

```python
len(name)
```

```python
print name[0:3]
```

```python
print name[3:6]
```

```python
print name[6:]
```

好在所有一切, 我都可以改成 unicode!

```python
name_unicode = name.decode('utf-8')
```

```python
print name_unicode[0]
```

```python
len(name_unicode)
```

```python
name_utf8 = name_unicode.encode('utf-8')
```

```python
print name_utf8[0]
```

## 小結論

* 把 Unicode 編碼成某個 byte codes 叫做 <b style="color:red;">encode</b>
* 把某種 byte codes 編碼還原為 Unicode 叫做 <b style="color:red;">decode</b>
